```python
import pandas as pd
```

```python
setups = {
    "moga": ["tuned"],
    "cma": ["default", "tuned"],
    "hype": ["default", "tuned"], 
    "ibea": ["default", "tuned"], 
    "moead": ["default", "tuned"], 
    "nsga": ["default", "tuned"], 
    "nsga3": ["default", "tuned"], 
    "sms": ["default", "tuned"], 
    "spea": ["default", "tuned"],
}
fes = ["2500", "10000", "40000"]
algos = ["cma", "hype", "ibea", "moead", "nsga", "nsga3", "sms", "spea"]
indicators = ["rpd", "eps", "igd"]
nobjs = ["2", "3", "5", "10"]
benchmarks = ["DTLZ", "WFG"]
problems = {
    "WFG": list(range(1,10)),
    "DTLZ": [2, 4, 5, 6, 7],
}
nvars = {
    "2":  [30, 40, 50],
    "3":  [30, 40, 50],
    "5":  [30, 40, 50],
    "10": [31, 41, 51],
}
seeds = list(range(25))
```

```python
data = [
    {
    "setup": setup,
    "FE": fe,
    "algo": algo,
    "indicator": indicator,
    "nobj": nobj,
    "problem": f"{bench}{problem}",
    "nvar": nvar,
    "seed": seeds,
    "value": pd.read_csv(f"{setup}-{fe}fes/{algo}/{algo}.{bench}{problem}.{nobj}.{nvar}.out_{indicator}", header=None)[0].tolist()
#     "file": f"{setup}-{fe}fes/{algo}/{algo}.{bench}{problem}.{nobj}.{nvar}.out_{indicator}"
    }
    for algo in algos
    for setup in setups[algo]
    for fe in fes
    for indicator in indicators
    for nobj in nobjs
    for bench in benchmarks
    for problem in problems[bench]
    for nvar in nvars[nobj]
#     for seed in seeds
]
```
