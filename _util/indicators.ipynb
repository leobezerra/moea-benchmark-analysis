```python
def rank_sum(df, columns=["algo"]):
    df_wide = df.pivot_table(
        index=["indicator", "problem", "nvar", "seed"],
        columns=columns, 
        values=["value"]
    )
    
    return df_wide.rank(axis=1).groupby("indicator").sum()

df_rs = df_errata.groupby(["setup","FE", "nobj"]).apply(rank_sum).droplevel(0, axis=1)
```

```python
df_rs
```

```python
for FE in [2500,10000,40000]:
    for indicator in ["rpd", "eps", "igd"]:
        idx = ("tuned", FE, 10, indicator)
        rs_diff = (df_rs.loc[idx] - df_rs.loc[idx].min())
        display(rs_diff.sort_values().to_frame().T)
```

- 2500: minimal changes according to IGD, but greatly improves according to RPD (3rd best ranked) and EPS
- 10000: minimal changes according to EPS and IGD, but greatly improves according to RPD (3rd best ranked)
- 40000: minimal changes according to IGD (surpasses CMA and SMS), but greatly improves according to RPD (4rd best ranked) and EPS
