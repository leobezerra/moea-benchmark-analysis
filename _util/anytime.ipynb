```python
import pandas as pd
```

# IBEA

```python
setups = ["default", "tuned"]
configs = {
    "default": [""],
    "tuned": ["2500", "10000", "40000"],
}
algos = ["ibea"]
indicators = ["rpd"]
nobjs = ["3"]
benchmarks = ["DTLZ", "WFG"]
problems = {
    "WFG": list(range(1,10)),
    "DTLZ": [2, 4, 5, 6, 7],
}
nvars = {
    "2":  [30, 40, 50],
    "3":  [30, 40, 50],
    "5":  [30, 40, 50],
}
seeds = list(range(1,26))
```

```python
data = [
    pd.read_csv(f"testing-anytime/{algo}{config}/{algo}.{bench}{problem}.{nobj}.{nvar}.{seed}.any", header=None, sep=" ")\
                .rename(columns={0: "FE", 1: "value"})\
                .set_index("FE").T\
                .assign(**{
                    "setup": setup,
                    "config": config,
                    "algo": algo,
                    "indicator": indicator,
                    "nobj": nobj,
                    "problem": f"{bench}{problem}",
                    "nvar": nvar,
                    "seed": seed,
                })
    for setup in setups
    for config in configs[setup]
    for algo in algos
    for indicator in indicators
    for nobj in nobjs
    for bench in benchmarks
    for problem in problems[bench]
    for nvar in nvars[nobj]
    for seed in seeds
]
```

```python
df = pd.concat(data)
df = df.set_index(["setup","config","algo","indicator","nobj","problem","nvar","seed"])
#df.to_csv("anytime-ibea.csv.gz", compression="gzip")
```
